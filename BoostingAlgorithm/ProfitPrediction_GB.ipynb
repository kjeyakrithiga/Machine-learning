Import Required python modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
import pickle
```

```python
dataset = pd.read_csv("../../Dataset/50_Startups_dataset.csv")
dataset
```

```python
dataset.columns
```

```python
dataset = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State',
       'Profit']]
dataset
```

OneHot encoding

```python
dataset = pd.get_dummies(dataset,dtype = int, drop_first=True)
dataset
```

```python
dataset.shape
dataset.describe()
```

```python
independent = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
dependent = dataset[['Profit']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,y_train,x_test,y_test
```

```python
#regressor = GradientBoostingRegressor(loss='absolute_error',random_state=0,criterion='friedman_mse') : 85
#regressor = GradientBoostingRegressor(loss='squared_error',random_state=0,criterion='friedman_mse') : 92
regressor = GradientBoostingRegressor(n_estimators = 50,loss='huber',random_state=0,criterion='squared_error')
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
y_pred.shape
```

```python
r_score=r2_score(y_pred,y_test)
r_score
```
