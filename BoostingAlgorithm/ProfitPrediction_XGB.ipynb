```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split 
import xgboost as xgb
from sklearn.metrics import r2_score
import pickle
```

```python
dataset = pd.read_csv("../../Dataset/50_Startups_dataset.csv")
dataset
```

```python
dataset.columns
```

```python
dataset = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State',
       'Profit']]
dataset
```

OneHot encoding

```python
dataset = pd.get_dummies(dataset,dtype = int, drop_first=True)
dataset
```

```python
dataset.shape
dataset.describe()
```

```python
independent = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
dependent = dataset[['Profit']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,y_train,x_test,y_test
```

```python
regressor = xgb.DMatrix(dataset)
num_round = 15
param = {'max_depth': 5, 'eta': 1, 'objective': 'reg:squarederror'}
param['nthread'] = 10
param['eval_metric'] = 'auc'
evals_result = {}
dtrain = xgb.DMatrix(x_train,label=y_train)
dtest = xgb.DMatrix(x_test,label=y_test)
#dtest_quantile = xgb.QuantileDMatrix(data=dtest, ref=dtrain)
eval_s = [(dtrain, "train"), (dtest, "eval")]
model = xgb.train(param, dtrain, num_round, eval_s,evals_result=evals_result)
```

```python
y_pred = model.predict(dtest)
y_pred
```

```python
y_pred.shape
```

```python
r_score=r2_score(y_pred,y_test)
r_score
```
