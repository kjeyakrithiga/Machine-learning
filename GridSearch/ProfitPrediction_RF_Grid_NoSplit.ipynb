```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
import pickle
```

```python
#dataset = pd.read_csv("../../Dataset/50_Startups_dataset.csv")
dataset = pd.read_csv("../../Dataset/insurance_pre.csv")
dataset
```

```python
dataset.columns
```

```python
dataset = dataset[['age', 'sex', 'bmi', 'children', 'smoker','charges']]
dataset
```

OneHot encoding

```python
dataset = pd.get_dummies(dataset,dtype = int, drop_first=True)
dataset
```

```python
dataset.shape
dataset.describe()
```

```python
independent = dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#x_train,y_train,x_test,y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent = sc.fit_transform(independent)
#x_test = sc.fit(x_test)
```

```python
param_grid = {'n_estimators':[10,50,100,200],'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'max_features':['auto','sqrt', 'log2']}
grid = GridSearchCV(RandomForestRegressor(),param_grid,refit=True,n_jobs = -1,verbose=3)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The R score value for the best parameter {}:".format(grid.best_params_))
table = pd.DataFrame.from_dict(re)
table
```
