Import Required python modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
import pickle
```

```python
dataset = pd.read_csv("../../Dataset/50_Startups_dataset.csv")
dataset
```

```python
dataset.columns
```

```python
dataset = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State',
       'Profit']]
dataset
```

OneHot encoding

```python
dataset = pd.get_dummies(dataset,dtype = int, drop_first=True)
dataset
```

```python
dataset.shape
dataset.describe()
```

```python
independent = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
dependent = dataset[['Profit']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,y_train,x_test,y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit(x_test)
```

```python
param_grid = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'gamma':['scale', 'auto'],'C':[10,100,1000,2000,3000]}
grid = GridSearchCV(SVR(),param_grid,refit=True,n_jobs = -1,verbose=3)
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print("The R score value for the best parameter {}:".format(grid.best_params_))
table = pd.DataFrame.from_dict(re)
table
```
