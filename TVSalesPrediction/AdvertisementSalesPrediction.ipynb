```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pickle
```

```python
dataset = pd.read_csv("../../Dataset/advertising.csv")
dataset
```

```python
dataset.shape
dataset.describe()
```

```python
x_points=np.array(dataset[['TV']])
y_points=np.array(dataset[['Sales']])
```

```python
plt.plot(x_points,y_points,'o')
plt.show()
```

```python
independent = dataset[['TV']]
dependent = dataset[['Sales']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,y_train,x_test,y_test
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bias = regressor.intercept_
bias
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
r_score=r2_score(y_pred,y_test)
r_score
```

```python
salary_predition_model = pickle.dump(regressor,open("../../Models/TVSalesPrediction_SLR.sav",'wb'))
```
